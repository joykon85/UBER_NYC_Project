# 1b. Add the borough information

```python
df_loc = pd.read_csv("taxi_zone_lookup.csv")
```

```python
df_loc
```

```python
merged_df = df.merge(df_loc[['LocationID', 'Borough']], left_on="PULocationID", right_on="LocationID", how="left")
```

```python
merged_df
```

```python
merged_df['Borough'].unique()
```

```python
borough_dfs = {borough: df_group for borough, df_group in merged_df.groupby("Borough")}
```

# Save and do seperately

```python
for borough, df_group in borough_dfs.items():
    # Example: Print borough name and first few rows
    print(f"Processing {borough}...")
    df_group.to_csv(f"{borough}.csv", index=False)
```
